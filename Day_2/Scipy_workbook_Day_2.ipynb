```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate
import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.arange(0,10)
y = np.exp(-x/3) # definig the y using exponential function , this is manual definition
```

```python
x
```

```python
plt.plot(x,y)
```

Interpolation - x and y are arrays of values used to approximate some function f: y = f(x). This class returns a function whose call method uses interpolation to find the value of new points.

Interpolation means finding new value of y wrt new x value . xo < xnew < x , if we are trying to find new x value which  lies in between the xo and x then it is called interpolation else it is extrapolation
- There are two types of iterpolation
1. Ploynomial Interpolation
2. Spline Interpolation

```python
# definining the interpolate function
f = interpolate.interp1d(x,y)
```

```python
f
```

```python
xnew = np.arange(0,9,0.1)
ynew = f(xnew) # using the interpolate function directly to define y points
plt.plot(xnew,ynew,'--',color ='r')
```

```python
# lets compare both the plots created using manually and interpolate function
plt.plot(x,y,'o',xnew,ynew,'-')
plt.show()
```

We can clearly observe that the interpolation points are matching maually defined function values

## Spline Interpolation-
Given the set of data points (x[i], y[i]) determine a smooth spline approximation of degree k on the interval xb <= x <= xe.

splev- Evaluate a B-spline or its derivatives.

Given the knots and coefficients of a B-spline representation, evaluate
the value of the smoothing polynomial and its derivatives.  This is a
wrapper around the FORTRAN routines splev and splder of FITPACK.

```python
from scipy.interpolate import splev,splrep
```

```python
x = np.linspace(0,10,10)
y = np.sin(x)
```

```python
plt.plot(x,y) # sharp curve 
plt.title('Linear Spline')
```

```python
# defining spline interpolation function
spl = splrep(x,y) # new points are stored in spl
```

```python
spl
```

```python
xnew = np.linspace(0,10,200)
ynew = splev(xnew,spl)
```

```python
plt.plot(xnew,ynew) # curve smoothening by interpolating
```

```python
# comparing both plots
plt.plot(x,y,'o',xnew,ynew,'-')
plt.title('Quadratric Spline')
plt.show()
```

Now we can see the curve turned out of smooth after converting the linear spline to Quadratic spline

## 2D- Interpolation

```python
# constructing a 2D grid
x = np.arange(-5.01,5.01,0.25)
y = np.arange(-5.01,5.01,0.25)
xx,yy = np.meshgrid(x,y)
```

```python
plt.plot(x,y)
```

```python
z =np.sin(xx**2 + yy**2)
z
```

```python
z.shape
```

```python
plt.plot(x,z[0,:],'ro-') # plotting only first row of z values with all columns
```

```python
f = interpolate.interp2d(x,y,z,kind = 'cubic')
```

```python
f
```

```python
# now use obtained interpolation result to plot
xnew = np.arange(-5.01,5.01,1e-2)
ynew = np.arange(-5.01,5.01,1e-2)
znew = f(xnew,ynew)
plt.plot(xnew,znew[0,:],'b-')
```

```python
# comparing the plots
plt.plot(x,z[0,:],'ro-',xnew,znew[0,:],'b-')
```
